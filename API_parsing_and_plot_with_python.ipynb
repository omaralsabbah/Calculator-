```python
import requests
response = requests.get('https://api.covid19api.com/summary').text
response
```

```python
import json
response_info = json.loads(response)
response_info['Countries']
```

```python
country_list=[{'Country':country_info['Country'],'Cases':country_info['TotalConfirmed']} for country_info in response_info['Countries']]
country_list
```

```python
import pandas as pd
df=pd.DataFrame(country_list)
df[df.Country=='United States of America']
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(40,5))
plt.bar(df['Country'],df['Cases'])

plt.xticks(rotation=90)
plt.show()
```
